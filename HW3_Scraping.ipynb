Your mission is to scrape the  [**CUNY Fall 2021 academic calendar** ](https://www.ccny.cuny.edu/registrar/fall) site using the requests library, beautiful soup, and pandas, just like in class. You should wind up with a pandas data frame where the index column is a Python date. There should be a column for “day of the week” with a variable label down and a column called text with the explanation.




```python
#Import necessary libraries

import requests as req
import bs4
from requests.models import Response
import pandas as pd

#Load the web paged and get the code of the required table
website = "https://www.ccny.cuny.edu/registrar/fall"
result = req.get(website)
content = result.text
soup = bs4.BeautifulSoup(content, "html.parser")

tables = soup.find_all('table')
table = tables[0]

#get the body of the table and the each row of the table
t_body = table.find_all('tbody')[0]
rows = [row for row in t_body.find_all('tr')]
```

```python
#Add each tow tag to the table that is related to it

dates = []
days = []
events = []

for row in rows[:]:
  date = row.findAll('td')[0].text
  day = row.findAll('td')[1].text
  event = row.findAll('td')[2].text

  dates.append(date)
  days.append(day)
  events.append(event)

dates_n = [date.strip() for date in dates]
days_n = [day.strip() for day in days]
events_n = [event.strip() for event in events]
#events_modify =[[event.replace(";\n\t\t\t", " ") for event in events_n]]
```

```python
df = pd.DataFrame(list(zip(dates_n, days_n, events_n)),columns=['Dates', 'Day of the week', 'Text with the explanation'])

df = df.reset_index(drop=True)

df.dtypes
```
